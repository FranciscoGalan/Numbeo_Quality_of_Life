```python
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.max_rows = 50
import re
```

# Data cleaning

```python
file_path = 'Quality_of_Life_Index.csv'
data = pd.read_csv(file_path)
```

```python
data.City[:50]
```

```python
data1 = data.copy()
```

```python
# Function to extract city
def extract_city(location):
    parts = location.split(',')
    city = parts[0].strip()
    if '(' in city:
        city = city.split('(')[0].strip()
    return city

# Function to extract country
def extract_country(location):
    parts = location.split(',')
    country = parts[-1].strip()
    return country
```

```python
# Extract country from the original 'City' column
data1.insert(2, 'Country', data1['City'].apply(extract_country))

# Extract city from the original 'City' column
data1['City'] = data1['City'].apply(extract_city)
```

Checking to see if all the values look good

# Save clean dataset

```python
data_clean = data1.copy()
data_clean.to_csv('Quality_of_Life_Index_clean.csv', index=False)
```
